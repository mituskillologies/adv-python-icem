```python
# 3. Find information of the players whose name starts with ‘A’.
cri[cri['name'].str.startswith('A')]
```

```python
# 4. Convert all the names in upper-case and replace it with original values.
cri['name'] = cri['name'].str.upper()
cri
```

```python
# 5. Find the addition of all the runs.
cri['runs'].sum()
```

```python
# 6. Print the name of player with highest runs.
cri[cri['runs']==cri['runs'].max()]['name']
```

```python
# 7. find  name of wicketkeeper
cri[cri['stumpings'] > 0]['name']
```

```python
# 8. Find average runs of the players who
# played 200+ matches
cri[cri['matches'] > 200]['runs'].mean()
```

```python
# 9. Name of the player with highest runs per
# match
rpm = max(cri['runs'] / cri['matches'])
cri[cri['runs'] / cri['matches'] == rpm]['name']
```

```python
# 10. Find names of the player with 50+ runs per
# match and 50+ catches
temp = cri['runs'] / cri['matches'] > 50
cri[temp & (cri['catches'] > 50)]['name']
```
