```python
stud = pd.read_csv('student.csv')
```

```python
plt.figure(figsize=(16,9))
plt.title('student performance')
plt.grid()
plt.xlabel('Student Name')
plt.ylabel('Marks')
plt.bar(stud['name'], stud['marks'],
        color='silver')
plt.savefig('students.png')
```

```python
plt.figure(figsize=(16,9))
plt.title('student performance')
plt.grid()
plt.ylabel('Student Name')
plt.xlabel('Marks')
plt.barh(stud['name'], stud['marks'],
        color='cyan')
```

#### Pie Chart

```python
plt.figure(figsize=(10,10))
plt.title('% share of students courses')
plt.pie(studs, labels=course, 
        autopct='%2.2f%%',shadow=True);
```
